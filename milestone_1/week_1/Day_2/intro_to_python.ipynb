### Strings: The Text Blocks
- In Python, a **String** is any sequence of characters wrapped in quotes (either `'single'` or `"double"`). Think of it as a literal piece of paper with words written on it.

#### Common String Operations
- **Concatenation**: Gluing strings together with `+`.

- **Repetition**: Repeating strings with `*`.

- **Methods**: Changing the text (e.g., `.upper()`, `.lower()`).

#### Example:

```python
# Basic String
greeting = "hello"
target = "world"

# Gluing them (The old way)
combined = greeting + " " + target + "!" 
print(combined.capitalize()) # Output: Hello world!

# Repeating
echo = "Hello! " * 3
print(echo) # Output: Hello! Hello! Hello!
```

#### F-Strings: The "Smart" Way

- Introduced in Python 3.6, F-Strings (formatted string literals) are the modern way to include variables inside text.

- The Context: Instead of cutting and pasting strings together like a ransom note, you create a template and use curly braces `{}` as placeholders.

#### Why use F-Strings?
- `Readability`: You can see the whole sentence at once.

- `Speed`: They are faster for the computer to process.

- `In-place Math`: You can actually run logic inside the braces.
#### Example:

```python
name = "Sarah"
score = 95
attempts = 3

# The F-String approach
# Note the 'f' prefix before the opening quote
message = f"Congrats {name}! You scored {score}% after {attempts} tries."

print(message) 
# Output: Congrats Sarah! You scored 95% after 3 tries.

# You can even do math inside them!
print(f"Next year, you will be {20 + 5} years old.")
```

### The Big Four Collections
#### 1. Lists: The Grocery List (Ordered & Changeable)
- A List is your go-to container. It keeps items in a specific order, and you can add, remove, or change them whenever you want.

- Syntax: Uses square brackets `[]`.

```python
# A list of fruit
fruits = ["apple", "banana", "cherry"]

fruits.append("orange")  # Adding an item
fruits[0] = "strawberry" # Changing the first item (Index starts at 0)

print(fruits) # ['strawberry', 'banana', 'cherry', 'orange']
```

```python
# Our starting list
fruits = ["strawberry", "banana", "cherry", "orange"]

# Pick one fruit to demonstrate methods
target = fruits[0]  # "strawberry"

print(f"Original: {target}")
print(f"Uppercase: {target.upper()}")           # STRAWBERRY
print(f"Capitalize: {target.capitalize()}")     # Strawberry
print(f"Replace: {target.replace('berry', ' blueberry')}") # straw blueberry
print(f"Count 'r': {target.count('r')}")        # 3
print(f"Start with 's'?: {target.startswith('s')}") # True

# Demonstrating Split and Join
csv_fruits = "apple,pear,grape"
split_list = csv_fruits.split(",")              # Turns string into List
print(f"Split List: {split_list}")

rejoined = " & ".join(fruits)                   # Turns List into String
print(f"Joined String: {rejoined}")

# Chaining methods
print(f"Chained: {fruits[1].upper().replace('A', '@')}") # B@N@N@
```

#### 2. Tuples: The Coordinates (Ordered but Unchangeable)
- A Tuple is like a List, but it is **immutable**—once you create it, you cannot change it. Use this for data that should stay "locked," like GPS coordinates or a birthdate.

- Syntax: Uses parentheses `()`.

```python
# GPS Coordinates (Latitude, Longitude)
location = (40.7128, 74.0060)

print(location[1])    # Accessing works fine
```

```python
# Our starting Tuple (Immutable)
location = (40.7128, 74.0060)

# 1. Accessing by Index
lat = location[0]
lon = location[1]
print(f"Latitude: {lat}, Longitude: {lon}")

# 2. Tuple Unpacking (The Pythonic Way)
# This assigns variables to each item in one line
y, x = location 
print(f"Unpacked -> Y: {y}, X: {x}")

# 3. Useful Methods
print(f"Length of tuple: {len(location)}")    # 2
print(f"Count occurrences of 40.7128: {location.count(40.7128)}") # 1
print(f"Find index of 74.006: {location.index(74.0060)}")        # 1

# 4. Membership Testing
is_present = 40.7128 in location
print(f"Is 40.7128 in the tuple? {is_present}")


# location[0] = 50.0  # This would cause an ERROR!
```

#### 3. Sets: The Guest List (Unique & Unordered)
- A Set is a collection where every item must be unique. If you try to add the same thing twice, it just ignores the second one. It doesn't keep items in any specific order.

- Syntax: Uses curly braces `{}`.

```python
# A set of unique User IDs
user_ids = {101, 102, 103, 101} # 101 is a duplicate

print(user_ids) # {101, 102, 103} (Duplicate is gone!)
```

```python
# Our starting Set
user_ids = {101, 102, 103, 101}

# 1. Basic Operations
user_ids.add(104)          # Add a new item
user_ids.remove(102)       # Remove an item
print(f"Updated Set: {user_ids}") 

# 2. Mathematical Set Logic (The "Venn Diagram" stuff)
active_users = {101, 103, 104}
admin_users = {104, 105, 106}

# Union: Everyone in both sets
all_users = active_users.union(admin_users) 
print(f"Union (All): {all_users}") # {101, 103, 104, 105, 106}

# Intersection: Only those who are in BOTH sets
admins_who_are_active = active_users.intersection(admin_users)
print(f"Intersection: {admins_who_are_active}") # {104}

# Difference: In active but NOT in admin
standard_users = active_users.difference(admin_users)
print(f"Difference: {standard_users}") # {101, 103}

# 3. Membership Testing (Super fast in Sets)
print(f"Is 101 present? {101 in active_users}") # True
```

#### 4. Dictionaries: The Contact List (Key-Value Pairs)
- A Dictionary (or dict) stores data in pairs: a Key and a Value. It’s like a real dictionary: you look up a "word" (key) to find its "definition" (value).

- Syntax: Uses curly braces with colons `{key: value}`.

```python
# A user profile
user = {
    "name": "Aris",
    "email": "aris@example.com",
    "level": 5
}

print(user["name"]) # Output: Aris
```

```python
# Our starting Dictionary
user = {
    "name": "Aris",
    "email": "aris@example.com",
    "level": 5
}

# 1. Accessing Data
print(f"User Name: {user['name']}") # Standard way
print(f"User Status: {user.get('status', 'Guest')}") # Safe way (prevents errors if key is missing)

# 2. Modifying & Adding Data
user["level"] = 6             # Updating an existing value
user["is_active"] = True       # Adding a new key-value pair
print(f"Updated User: {user}")

# 3. Viewing Keys, Values, and Items
print(f"Keys: {list(user.keys())}")      # ['name', 'email', 'level', 'is_active']
print(f"Values: {list(user.values())}")  # ['Aris', 'aris@example.com', 6, True]

# 4. Removing Data
removed_value = user.pop("email")
print(f"Removed Email: {removed_value}")
print(f"Final Dict: {user}")

# 5. Iterating (Common for Loops)
for key, value in user.items():
    print(f"{key.capitalize()}: {value}")
```

### Control Flow & Logic
- Control flow is how you tell the computer to make decisions or repeat actions. It’s like giving directions: "If it’s raining, take an umbrella; otherwise, enjoy the sunshine."
- Python uses `if`, `elif`, and `else` for decision-making, and loops like `for` and `while` for repetition.

#### 1. Conditional Logic (`if`, `elif`, `else`):
- Think of this as a fork in the road. You ask a question (a condition), and if the answer is True, you take that path.
#### Example:

```python
score = 85

if score >= 90:
    print("Grade: A - Excellent!")
elif score >= 80:
    print("Grade: B - Good job!")
elif score >= 70:
    print("Grade: C - Keep trying.")
else:
    print("Grade: F - Let's study more.")
```

#### 2. Logical Operators (`and`, `or`, `not`):
Sometimes one condition isn't enough. We use logical operators to combine them.

- `and`: Both must be true.

- `or`: At least one must be true.

- `not`: Flips the result (True becomes False).

```python
has_ticket = True
is_dressed_up = False

if has_ticket and is_dressed_up:
    print("Welcome to the Gala!")
elif has_ticket or is_dressed_up:
    print("You're halfway there, but you can't enter yet.")
else:
    print("Access Denied.")
```

#### 3. Loops: Doing Work Repeatedly
Loops are for when you need to do the same task multiple times without writing the code over and over.
- The `for` loop (Iterating):
Best for when you know how many times you want to run (like through a list).

```python
fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
    print(f"I am eating a {fruit}")
```

- The `while` loop (Conditional): Runs as long as a condition remains True. Be careful of "Infinite Loops"!

```python
countdown = 3

while countdown > 0:
    print(f"Launching in {countdown}...")
    countdown -= 1  # Important: change the value so the loop eventually ends!

print("Blast off!")
```

The "Teacher's Secret": `break` and `continue`:

- `break`: Stop the loop immediately (emergency exit).

- `continue`: Skip the rest of this round and jump to the next one.

```python
for i in range(1, 6):
    if i == 3:
        continue  # Skip 3
    if i == 5:
        break     # Stop entirely at 5
    print(f"Number: {i}")
```

#### List Comprehensions: The One-Liner Magic

Usually, to create a list based on another list, you would:

- 1. Create an empty list.

- 2. Start a for loop.

- 3. Apply logic (like math or an if statement).

- 4. append the result.

However, using the `list comprehension` does all of that in one bracket [].

```python
numbers = [1, 2, 3, 4, 5, 6]

# --- THE LONG WAY (4 lines) ---
squares = []
for n in numbers:
    if n % 2 == 0:
        squares.append(n**2)

print(f"Long way: {squares}") # [4, 16, 36]


# --- THE LIST COMPREHENSION WAY (1 line) ---
# Formula: [expression for item in iterable if condition]
fast_squares = [n**2 for n in numbers if n % 2 == 0]

print(f"Short way: {fast_squares}") # [4, 16, 36]
```

### Functions & Modular:

#### The Anatomy of a Function: 
1. `def`: The keyword that tells Python you are "defining" a function.

2. `Name`: How you call the function later (e.g., calculate_tax).

### Exceptional Handling: The Safety Net

When you run code, things will eventually go wrong: a file is missing, a user enters a letter instead of a number, or a server is down. In standard code, these errors (Exceptions) are like a gymnast falling—the performance stops immediately. Exception Handling is the safety net that catches the fall so the show can go on.

#### The `try...except` Concept:

- Instead of just running risky code, we "try" to run it. If it fails, we "catch" the error and tell the computer how to respond gracefully instead of crashing.

- The Key Keywords:
    - `try`: The block of code you want to run (the risky part).

    - `except`: The code that runs only if an error happens.

    - `else`: (Optional) Runs only if no errors happened.

    - `finally`: (Optional) Runs no matter what (used for cleanup).

```python
def divide_numbers():
    try:
        # Get user input
        num1 = float(input("Enter a number: "))
        num2 = float(input("Enter another number: "))
        
        # Perform the calculation
        result = num1 / num2
        
    except ZeroDivisionError:
        print("Error: You cannot divide by zero!")
    except ValueError:
        print("Error: Please enter valid numbers, not text.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        
    else:
        # This only runs if the 'try' block was successful
        print(f"The result is: {result}")
        
    finally:
        # This always runs, perfect for closing files or logging
        print("Calculation attempt finished.")

# Run the function
divide_numbers()
```

#### File Handling: Working with Files
- Python makes it easy to read from and write to files, which is essential for tasks like data analysis, logging, or saving user input.
- Working with files always follows three steps:

    1. Open the file.

    2. Act (Read the text or Write new text).

    3. Close the file (to save memory and prevent data corruption).

The Pythonic Way: We use the `with` keyword. This acts like an "Auto-Close" feature. Even if your code crashes, Python ensures the file is safely closed.


```python
# 1. WRITING to a file ('w' mode)
# If the file doesn't exist, Python creates it. 
# WARNING: 'w' overwrites everything in the file!
with open("session_notes.txt", "w") as file:
    file.write("Python Session 2026\n")
    file.write("Topic: File Handling\n")
    file.write("Status: In Progress")

print("File created and data written successfully!")

# 2. READING from a file ('r' mode)
try:
    with open("session_notes.txt", "r") as file:
        content = file.read()
        print("\n--- File Content ---")
        print(content)
except FileNotFoundError:
    print("Error: The file was not found.")

# 3. APPENDING to a file ('a' mode)
# Use 'a' to add new lines without deleting old ones.
with open("session_notes.txt", "a") as file:
    file.write("\nUpdate: File handling is easy!")
```

### Decorators: Functions as Wrappers
- A decorator is a function that takes another function as an input, adds some extra "logic" around it, and then returns it. It allows you to "inject" code into existing functions without actually changing their internal code

Example:

```python
# 1. Defining the Decorator (The Wrapping Paper)
def my_decorator(func):
    def wrapper():
        print("--- Initialization: System Check Starting ---")
        func()  # This is the "Gift" being opened
        print("--- Task Completed: Powering Down ---")
    return wrapper

# 2. Using the Decorator with the '@' symbol
@my_decorator
def launch_rocket():
    print("Rocket is launching into space!")

# 3. Running the function
launch_rocket()
```

- **Breaking Down the Syntax**:
    - `@my_decorator`: This is "Syntactic Sugar." It’s a shortcut for saying `launch_rocket = my_decorator(launch_rocket)`.

    - The Nested `wrapper()`: This is where the magic happens. We put the code we want to run before and after the original function here.

    - `return wrapper`: The decorator hands back the "wrapped" version of the function to be used later.

#### OOPs (Object-Oriented Programming)
- To explain OOP (Object-Oriented Programming), use the "Architect and the House" analogy. In procedural programming (what we’ve done so far), we focus on the steps (the recipe). In OOP, we focus on the objects (the finished dish).

##### The Concept: Blueprints vs. Real Things:

- **Class**: This is the **Blueprint**. It’s the drawing of a house that says it should have a door, windows, and a color. The blueprint itself isn't a house; you can't live in it.

- **Object (Instance)**: This is the **Actual House** built from the blueprint. You can build 100 different houses from one blueprint—some might be blue, some might be red—but they all follow the same structure.

##### The Code: Creating a Template
Let's build a "blueprint" for a Car.

```python
# 1. Defining the Blueprint (Class)
class Car:
    # The "Constructor" - It sets up the starting data
    def __init__(self, brand, color, speed):
        self.brand = brand      # Attribute
        self.color = color      # Attribute
        self.speed = speed      # Attribute

    # A "Method" - This is an action the object can do
    def drive(self):
        print(f"The {self.color} {self.brand} is zooming at {self.speed} km/h!")

# 2. Building Objects (Instances) from the Blueprint
car1 = Car("Tesla", "White", 120)
car2 = Car("Ford", "Red", 100)

# 3. Interacting with the Objects
print(car1.brand) # Accessing data (Attribute)
car2.drive()      # Running an action (Method)
```

##### The "Big Four" Pillars: 

1. **Encapsulation**: Keeping the data (attributes) and the tools (methods) in one locked box.

- It is about bundling data (variables) and behavior (functions) into a single unit (a Class) and hiding the messy details.
- Take the example of a car: We hide the internal state of the car so the driver can't accidentally set the fuel level to a negative number or the speed to a million.

```python
class Car:
    def __init__(self):
        self.__fuel_level = 100  # Private attribute (hidden from user)

    def drive(self):
        if self.__fuel_level > 0:
            self.__fuel_level -= 10
            print(f"Driving... Fuel left: {self.__fuel_level}%")
        else:
            print("Out of fuel!")

my_car = Car()
my_car.drive()
# my_car.__fuel_level = 5000  <-- This would cause an error!
```

2. **Inheritance**: Creating a "Child" class that steals features from a "Parent" class (e.g., an `ElectricCar` class inheriting from the `Car` class).
- Why write the same code twice? If you have a "Parent" class with general features, a "Child" class can "inherit" those features and then add its own unique ones.
- Example: If you have a `Car` class with attributes like `make`, `model`, and methods like `start_engine()`, you can create an `ElectricCar` class that inherits all those features and then adds its own, like `charge()`.

```python
class Car: # Parent Class
    def start_engine(self):
        print("Vroom! Engine started.")

class ElectricCar(Car): # Child Class
    def charge(self):
        print("Charging the battery...")

tesla = ElectricCar()
tesla.start_engine()  # Inherited from Car
tesla.charge()        # Unique to ElectricCar
```

3. **Abstraction**: The "Strict" Interface
Abstraction is about showing only the essential "buttons" to the user while hiding the complex background logic. Using the ABC module, we also turn this into a rule that all cars must follow.

- The Blueprint (The "What"): We define that every vehicle must have a fuel_type, but we don't define it in the blueprint because every car is different.

- The Interface (The "Button"): We provide a single push_button() method. The driver doesn't need to see the "private" steps like checking the battery or igniting the engine.

- The Hidden Logic (The "How"): The complex internal methods are hidden (using __) so the driver can't accidentally trigger them in the wrong order.

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def fuel_type(self):
        """CONTRACT: Every child class MUST define this."""
        pass

    def push_button(self):
        """INTERFACE: The only button the user needs to see."""
        self.__check_battery()   # Hidden complex step 1
        self.__ignite_engine()   # Hidden complex step 2
        print(f"The {self.fuel_type()} engine is ready!")

    def __check_battery(self):   # PRIVATE: Hidden under the hood
        print("Checking battery...")

    def __ignite_engine(self):   # PRIVATE: Hidden under the hood
        print("Igniting...")

class PetrolCar(Vehicle):
    def fuel_type(self):
        return "Petrol"

my_car = PetrolCar()
my_car.push_button()
```

4. **Polymorphism**: Different objects can respond to the same command in their own way (e.g., a `Dog` and `Cat` both have a `make_sound()` method, but one barks and the other meows).
- Polymorphism allows us to use the same method name for different types of objects, and each object can respond in its own way. This makes our code more flexible and easier to maintain.
- Example: Different types of cars respond to the "Accelerate" command differently. A Sports Car roars, while an Electric Car hums silently.

```python
class SportsCar:
    def accelerate(self):
        return "ROAR! 0 to 100 in 3 seconds."

class ElectricCar:
    def accelerate(self):
        return "Whirrr... 0 to 100 in 2 seconds."

# The same 'accelerate' command works on a list of different cars
garage = [SportsCar(), ElectricCar()]

for car in garage:
    print(car.accelerate())
```

### Virtual Environments:

Imagine you are an artist. For one project, you need messy oil paints. For another, you need clean, delicate watercolors. If you use the same desk for both at the same time, the oil will ruin the watercolors. A Virtual Environment is like having a separate, clean studio for every project.

#### Why do we need it?

By default, Python installs everything in one big "global" folder. This is bad because:

- **Version Conflicts**: Project A might need "Library Version 1.0," but Project B needs "Library Version 2.0."

- **Messiness**: Your computer gets cluttered with hundreds of libraries you only used once.

- **Portability**: It’s hard to share your project with others if you don't know exactly which libraries it needs.

#### Steps to Create a Virtual Environment:

- **Step 1: Create the Environment**

This creates a folder (usually named `.venv` or `env`) that contains a copy of the Python interpreter just for this project.

    - Terminal (Windows/Mac/Linux):

```python
# syntax: python -m venv <name_of_environment>
python -m venv .venv
```

- **Step 2: Activate the Environment**

This tells your computer: "Stop using the global Python and start using the one inside this folder."

Terminal (Windows):


```python
.\env\Scripts\activate
```

Terminal (Mac/Linux):

```python
source env/bin/activate
```

- **Step 3: Install Libraries**

Now that your environment is active, you can install libraries using `pip install library_name`. These libraries will only be available in this virtual environment, keeping your projects clean and organized.

```python
pip install requests
```

- **Step 4: Deactivate the Environment**

When you're done working, you can deactivate the environment to return to your global Python setup.

```python
deactivate
```

- **Bonus Tip: Using `requirements.txt`**

When you want to share your project, you can create a `requirements.txt` file that lists all the libraries and their versions. This way, anyone can set up the same environment by running `pip install -r requirements.txt`.

# Create the list
pip freeze > requirements.txt

# Someone else can install everything from that list using:
pip install -r requirements.txt
