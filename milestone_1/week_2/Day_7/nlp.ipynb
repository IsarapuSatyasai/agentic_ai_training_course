## 3. Tokenization and Basic Terminologies
Tokenization is the first and most important step in NLP — it breaks text into individual tokens (words, punctuation, numbers). This cell shows how spaCy tokenizes text and extracts useful attributes like lemma, POS tag, and dependency.

```python
text = "Apple is looking to buy a U.K. startup for $1 billion in 2026. Elon Musk leads xAI."

doc = nlp(text)

print("Original Text:", text)
print("\n" + "="*60)
print("Tokenization Output")
print("="*60)

for token in doc:
    print(f"{token.text:15} | Lemma: {token.lemma_:12} | POS: {token.pos_:8} | "
          f"Tag: {token.tag_:8} | Dep: {token.dep_}")
```

## 4. Tokenization Practicals + Sentence Segmentation
This cell demonstrates sentence segmentation (splitting text into sentences) and shows additional token attributes like `is_alph`a, `is_stop`, and `is_punct`. Useful for understanding how `spaCy` handles real multi-sentence text.

```python
multi_text = """Natural Language Processing is fascinating. 
Apple Inc. is buying a startup in London. 
Will it rain today? Let's check the weather!"""

doc = nlp(multi_text)

print("Sentences:")
for i, sent in enumerate(doc.sents, 1):
    print(f"{i}. {sent.text.strip()}")

print("\nTokens with attributes:")
for token in doc:
    if not token.is_space:
        print(f"Token: {token.text:12} | Alpha: {token.is_alpha} | "
              f"Stopword: {token.is_stop} | Punct: {token.is_punct}")
```

## 5. Text Preprocessing: Stemming vs Lemmatization

It shows Stemming (using NLTK) vs Lemmatization (using spaCy) on the same set of words. This helps you clearly understand the difference between the two most important text normalization techniques.

```python
print("Stemming vs Lemmatization Comparison")
print("="*70)

stemmer = PorterStemmer()
snowball = SnowballStemmer("english")

words = ["running", "runs", "ran", "easily", "easy", "fairly", "fairness",
         "computing", "computers", "computed", "flies", "flying", "better", "best"]

doc = nlp(" ".join(words))

print(f"{'Word':12} {'Porter Stem':12} {'Snowball Stem':15} {'spaCy Lemma':12}")
print("-" * 65)

for token in doc:
    word = token.text
    print(f"{word:12} {stemmer.stem(word):12} {snowball.stem(word):15} {token.lemma_:12}")
```

## 6. Stopwords Removal
Stopwords are common words (like "the", "is", "a") that usually add little value in many NLP tasks. This cell shows how to view spaCy’s default stopwords and how to remove them from text.

```python
print("Total stopwords in spaCy:", len(nlp.Defaults.stop_words))
print("Sample stopwords:", list(nlp.Defaults.stop_words)[:15])

# Example
text = "This is a very good example of stopword removal in Natural Language Processing."
doc = nlp(text)

clean_tokens = [token.text for token in doc if not token.is_stop and not token.is_punct]
print("\nAfter removing stopwords:", " ".join(clean_tokens))
```

## 7. Parts of Speech (POS) Tagging
POS tagging assigns grammatical categories (Noun, Verb, Adjective, etc.) to each token. This is very useful for feature engineering and understanding sentence structure.

```python
text = "Apple is buying a U.K. startup for $1 billion in London."
doc = nlp(text)

print(f"{'Token':12} {'POS':8} Explanation")
print("-" * 50)
for token in doc:
    print(f"{token.text:12} {token.pos_:8} {spacy.explain(token.pos_)}")
```

## 8. Named Entity Recognition (NER)
NER identifies real-world objects like organizations, persons, dates, money, locations, etc. This is one of the most powerful features of spaCy.

```python
text = """Apple Inc. is planning to acquire a startup in London, U.K. 
for $1.2 billion on March 15, 2026. Tim Cook met with Elon Musk last week in California."""

doc = nlp(text)

print("Named Entities Detected:")
print("="*50)
for ent in doc.ents:
    print(f"{ent.text:25} → {ent.label_:12} ({spacy.explain(ent.label_)})")
```

## 9. Complete Text Analysis Function

```python
def analyze_text(text):
    doc = nlp(text)
    
    print("TOKENS:", [token.text for token in doc])
    print("\nLEMMAS (non-stop):", [token.lemma_ for token in doc if not token.is_stop])
    
    print("\nPOS Tags:")
    for token in doc:
        if not token.is_punct:
            print(f"  {token.text:12} → {token.pos_} ({spacy.explain(token.pos_)})")
    
    print("\nNAMED ENTITIES:")
    for ent in doc.ents:
        print(f"  {ent.text:20} → {ent.label_} ({spacy.explain(ent.label_)})")

# Test
analyze_text("Sundar Pichai visited Hyderabad in January 2025 and announced new AI initiatives.")
```
