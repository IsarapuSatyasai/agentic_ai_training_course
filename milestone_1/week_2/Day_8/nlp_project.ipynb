# TextInsight – A Modular File-to-Report NLP Pipeline

This project breaks down a raw text file into its linguistic components and exports a structured analytical report. Below is the code split into logical modules for better understanding.

## 1. Library Imports and Model Initialization

In this snippet, we import `spacy` and load the pre-trained language model. Loading the model at the top ensures it is ready for use across all functions.

```python
import spacy

# Load the small English model pipeline
# This model includes the tokenizer, lemmatizer, tagger, and entity recognizer
nlp = spacy.load("en_core_web_sm")

print("NLP Engine Ready.")
```

## 2. The File Reader Function
This function handles the Input part of the project. It uses a context manager (`with`) to safely open the file and read the raw text into a Python string.

```python
def read_raw_data(file_path):
    """Opens a file and returns the string content."""
    try:
        with open(file_path, "r", encoding="utf-8") as f:
            content = f.read()
        return content
    except FileNotFoundError:
        return "Error: The input file was not found."
```

## 3. The NLP Processing Core
This is the heart of the project. It takes the raw string, runs it through the nlp pipeline, and iterates through every token to extract its attributes.

```python
def extract_linguistic_features(text):
    """Processes text and returns a list of formatted strings."""
    doc = nlp(text)
    report_rows = []

    # Create Header
    report_rows.append(f"{'TOKEN':<15} | {'LEMMA':<15} | {'POS':<8} | {'ENTITY'}")
    report_rows.append("-" * 55)

    # Process each word (Tokenization, Lemmatization, POS, NER)
    for token in doc:
        # Get NER label if it exists, else use 'O' (Outside)
        label = token.ent_type_ if token.ent_type_ else "O"
        
        row = f"{token.text:<15} | {token.lemma_:<15} | {token.pos_:<8} | {label}"
        report_rows.append(row)
        
    return "\n".join(report_rows)
```

## 4. The File Exporter Function
This function handles the Output. It takes the structured string we created in the previous step and saves it to a new text file.

```python
def export_report(data, output_path):
    """Writes the processed report string to a text file."""
    with open(output_path, "w", encoding="utf-8") as f:
        f.write(data)
    print(f"Report successfully generated: {output_path}")
```

## 5. Main Execution Logic
This snippet ties all the functions together into a sequence. This is the part of the script you actually run.

```python
# 1. Define Paths
input_file = "raw.txt"
output_file = "nlp_report.txt"

# 2. Execute Pipeline
raw_string = read_raw_data(input_file)

if "Error" not in raw_string:
    # Transform raw text into insights
    processed_report = extract_linguistic_features(raw_string)
    
    # Save insights to the output file
    export_report(processed_report, output_file)
else:
    print(raw_string)
```

### **Summary of Concepts Used**

- **Tokenization**: Splitting the text into individual words and punctuation.

- **Lemmatization**: Reducing words like "bought" or "buying" to the base form "buy."

- **POS (Part-of-Speech)**: Identifying words as Nouns, Verbs, or Adjectives.

- **NER (Named Entity Recognition)**: Detecting names of people, companies, or locations.

- **File I/O**: Reading from and writing to external .txt files.
