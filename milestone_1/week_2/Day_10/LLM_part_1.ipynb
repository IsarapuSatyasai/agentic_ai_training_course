````python
import os
import time
from dotenv import load_dotenv
from huggingface_hub import InferenceClient

# Load API Key
load_dotenv()

# ========================= CONFIG =========================
MODEL_NAME = "Qwen/Qwen2.5-3B-Instruct"
MAX_NEW_TOKENS = 300

# ========================= CLIENT SETUP =========================
hf_token = os.getenv("HF_TOKEN")
if not hf_token or hf_token == "hf_your_actual_huggingface_token_here":
    print("Error: Please set your HF_TOKEN in the .env file")
    exit(1)

client = InferenceClient(model=MODEL_NAME, token=hf_token)

print("LLM Demo")
print("Agentic AI Course")
print("-" * 30)

# ====================== HELPER FUNCTIONS ======================
def count_tokens(text: str) -> int:
    import tiktoken
    enc = tiktoken.get_encoding("cl100k_base")
    return len(enc.encode(text))

def generate(prompt: str, temperature: float = 0.7, max_new_tokens: int = MAX_NEW_TOKENS):
    start_time = time.time()
    
    response = client.text_generation(
        prompt=prompt,
        max_new_tokens=max_new_tokens,
        temperature=temperature,
        top_p=0.9,
        stream=False
    )
    
    time_taken = time.time() - start_time
    input_tokens = count_tokens(prompt)
    output_tokens = count_tokens(response)
    
    return {
        "response": response.strip(),
        "time_taken": round(time_taken, 2),
        "input_tokens": input_tokens,
        "output_tokens": output_tokens,
        "total_tokens": input_tokens + output_tokens
    }

# ====================== 1. TEMPERATURE EXPERIMENT ======================
print("1. Temperature and Probabilistic Generation")

prompt = "Write a short creative story (one paragraph) about a robot who finds a glowing flower in a desert."

temperatures = [0.1, 0.7, 1.3]

for temp in temperatures:
    print(f"\nTemperature = {temp}")
    result = generate(prompt, temperature=temp, max_new_tokens=150)
    print(f"Time: {result['time_taken']}s | Total Tokens: {result['total_tokens']}")
    print(result['response'])
    print("-" * 85)

# ====================== 2. TOKENIZATION ======================
print("\n2. Tokenization Explorer")

texts = [
    "Hello, how are you today?",
    "Antidisestablishmentarianism",
    "The quick brown fox jumps over the lazy dog.",
    "Artificial Intelligence is evolving rapidly in 2026!"
]

for text in texts:
    tokens = count_tokens(text)
    print(f"Text: {text}")
    print(f"Tokens: {tokens}")
    print("-" * 60)

# ====================== 3. LOST IN THE MIDDLE ======================
print("\n3. Lost in the Middle Problem")

filler = "This is some unimportant background information. " * 70
secret = "VERY IMPORTANT: The secret password is PHOENIX-9247-ALPHA. "
test_prompt = filler + secret + "\n\nWhat is the secret password?"

result = generate(test_prompt, temperature=0.1, max_new_tokens=100)

print("Model Response:")
print(result['response'])
````
