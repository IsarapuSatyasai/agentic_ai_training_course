# Live Exercise: From Weak Prompt → Production-Grade Prompt

**Goal:**  
Transform a simple weak prompt into a strong, professional prompt ready for API and AI Agents.

**Task:**  
"Write a marketing email for my new productivity app."

## Step 1: The Weak Prompt (What most people do)

```python
from openai import OpenAI

client = OpenAI()

weak_prompt = "Write a marketing email for my new productivity app."

response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "user", "content": weak_prompt}
    ],
    temperature=0.7
)

print(response.choices[0].message.content)
```

## Problems with the Weak Prompt
- No clear role for the AI
- No context about the product
- No target audience
- No structure or tone instructions
- Output is random and inconsistent

## Step 2: Building a Strong Prompt using COSTAR

### Final Strong System Prompt (Best Version)

```python
system_prompt = """You are an expert direct-response copywriter with 10+ years of experience writing high-converting marketing emails for SaaS productivity tools.

Your writing style is clear, exciting, benefit-focused, and professional but friendly.
Always use positive language and focus on how the product improves the user's life."""

user_prompt = """Create a marketing email for "FocusFlow" - a new productivity app.

Product Details:
- Helps users focus deeply and reduce distractions
- Features: Pomodoro timer, smart task planner, AI distraction blocker
- Target users: Students, freelancers, and remote workers
- Pricing: $9/month after 14-day free trial

Use COSTAR method to create the email.
"""

full_prompt = system_prompt + "\n\n" + user_prompt
```

## Step 3: Adding Clear Response Format (Very Important for Agents)

```python
structured_prompt = """Create a high-converting marketing email for FocusFlow.

Product Details:
- Helps users focus deeply and reduce distractions
- Features: Pomodoro timer, smart task planner, AI distraction blocker
- Target users: Students, freelancers, and remote workers
- Pricing: $9/month after 14-day free trial

Return your response in this exact JSON format:

{
  "subject_line": "...",
  "preview_text": "...",
  "email_body": "... (use markdown formatting)",
  "cta_button_text": "...",
  "tone": "friendly and exciting"
}
"""
```

## Step 4: Final API Call with Strong Prompt

```python
response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": structured_prompt}
    ],
    temperature=0.5,      # Lower for more consistent output
    response_format={"type": "json_object"}   # Force JSON output
)

import json
result = json.loads(response.choices[0].message.content)
print(json.dumps(result, indent=2))
```

## Step 5: Comparison

**Weak Prompt Result:** Random, average quality, hard to use
**Strong Prompt Result:** Professional, structured, consistent, ready to use

This is the difference between normal ChatGPT user and Professional AI Agent Engineer!
