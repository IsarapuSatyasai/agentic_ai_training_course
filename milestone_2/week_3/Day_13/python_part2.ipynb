```python
# Example - 2
def process_scores(names: list[str], scores: list[int]) -> dict[str, int]:
  return dict(zip(names, scores))

# Execute
students = ['Ana', 'John']
grades = [92, 85]

process_scores(students, grades)
```

# Exception Handling

```python
def div(a : int, b : int) -> float:
  try:
    return a/b
  except ZeroDivisionError:
    return 'You Cannot divide a number by zero!'

# Usage
div(2,0)
```

## Decorators

```python
def decorate(func, a, b):
  def wrapper(a, b):
    print('Starting...')
    func(a, b)
    print('Ending..')
  return wrapper

@decorate
def display_msg():
  print("Welcome!")


display_msg()
```

```python
# *args
def sum(*args):
  total = 0
  for i in args:
    total += i
  return total

sum(5,1)


# *kwargs

def usernames(*kwargs):
  print(kwargs)
  for key, value in kwargs:
    print((key, value))

usernames(('john', 25), ('elisa', 27))
```
