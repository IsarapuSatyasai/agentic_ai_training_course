## Building a Mini Research & Analysis Agent

**Objective**: Create an intelligent agent that researches and analyzes "The Current State of AI Agents in 2026" and generates a professional structured report.

**Techniques Used**:
- Strong System Prompt (COSTAR style)
- Task Decomposition
- Chain of Thought (CoT) Reasoning
- Structured JSON Output
- Support for both OpenAI and Hugging Face

### 2. Import Libraries & Load Configuration

```python
import os
import json
from dotenv import load_dotenv
from openai import OpenAI
from huggingface_hub import InferenceClient

load_dotenv()

# Configuration
PROVIDER = "openai"          # Change to "huggingface" to switch provider

# OpenAI Settings
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")
OPENAI_MODEL = "gpt-4o-mini"

# Hugging Face Settings
HF_API_KEY = os.getenv("HUGGINGFACE_API_KEY")
HF_MODEL = "Qwen/Qwen2.5-32B-Instruct"

TEMPERATURE = 0.7
```

### 3. Define System and User Prompts

```python
SYSTEM_PROMPT = """You are an expert AI Research Analyst and Technical Writer specializing in AI Agents and Autonomous Systems.
You are highly analytical, objective, and skilled at structured reporting.
Always think step-by-step and provide well-reasoned insights."""

USER_PROMPT = """Conduct a comprehensive analysis on the current state of AI Agents in 2026.

Use Chain of Thought reasoning. Break down the task into clear steps.

Return your final response in this exact JSON format:

{
  "title": "Professional report title",
  "executive_summary": "2-3 sentence summary",
  "key_trends": ["trend 1", "trend 2", "trend 3"],
  "major_developments": [
    {
      "area": "Area name",
      "description": "What happened",
      "impact": "Why it matters"
    }
  ],
  "challenges": ["challenge 1", "challenge 2"],
  "future_outlook": "Prediction for next 12-18 months",
  "conclusion": "Final thoughts"
}

Think step by step before generating the final JSON."""
```

### 4. Define Functions for OpenAI and Hugging Face

```python
def get_openai_response():
    """Get response from OpenAI"""
    client = OpenAI(api_key=OPENAI_API_KEY)
    
    response = client.chat.completions.create(
        model=OPENAI_MODEL,
        messages=[
            {"role": "system", "content": SYSTEM_PROMPT},
            {"role": "user", "content": USER_PROMPT}
        ],
        temperature=TEMPERATURE,
        response_format={"type": "json_object"}
    )
    return response.choices[0].message.content


def get_huggingface_response():
    """Get response from Hugging Face Inference API"""
    client = InferenceClient(model=HF_MODEL, token=HF_API_KEY)
    
    prompt = f"<|im_start|>system\n{SYSTEM_PROMPT}<|im_end|>\n<|im_start|>user\n{USER_PROMPT}<|im_end|>\n<|im_start|>assistant\n"
    
    response = client.text_generation(
        prompt,
        max_new_tokens=2000,
        temperature=TEMPERATURE,
        stop=["<|im_end|>"]
    )
    return response
```

### 5. Run the Mini Agent

```python
print("=" * 70)
print("MINI RESEARCH & ANALYSIS AGENT")
print("Topic: Current State of AI Agents in 2026")
print("=" * 70)
print(f"Provider : {PROVIDER.upper()}")
print(f"Model    : {OPENAI_MODEL if PROVIDER == 'openai' else HF_MODEL}")
print("-" * 70)
print("Agent is thinking step by step...\n")

try:
    if PROVIDER == "openai":
        result = get_openai_response()
    else:
        result = get_huggingface_response()
    
    report = json.loads(result)
    
    print("\n" + "=" * 70)
    print("REPORT GENERATED SUCCESSFULLY")
    print("=" * 70)
    
    print(f"\nTitle: {report['title']}\n")
    print("Executive Summary:")
    print(report['executive_summary'])
    
    print("\nKey Trends:")
    for trend in report.get('key_trends', []):
        print(f"  • {trend}")
    
    print("\n" + "-" * 50)
    print("FULL REPORT (JSON)")
    print("-" * 50)
    print(json.dumps(report, indent=2))
    
except Exception as e:
    print(f"\nError: {e}")
```
